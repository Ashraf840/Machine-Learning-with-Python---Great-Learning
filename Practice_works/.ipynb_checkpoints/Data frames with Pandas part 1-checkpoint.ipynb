### Convert the "START_DATE" & "END_DATE" into Datetime-format

Doc: [**strftime() and strptime() Format Codes**](https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior)

```python
df.head()
```

```python
# Manual Conversion (START_DATE)
# Convert to datetime-format safely using errors="coerce" param, provide format thus it'll get direction
pd.to_datetime(df["START_DATE"], errors="coerce", format="%m/%d/%Y %H:%M")
```

```python
# Manual Conversion (START_DATE)
# Convert to datetime-format safely, provide format thus it'll get direction
pd.to_datetime(df["END_DATE"], errors="coerce", format="%m/%d/%Y %H:%M")
```

```python
# Datetime converter function, permanently
def datetime_converter(dframe):
    for dcol in dframe.columns:
        df[dcol] = pd.to_datetime(df[dcol], errors="coerce", format="%m/%d/%Y %H:%M")

datetime_converter(df[["START_DATE","END_DATE"]])
```

### Filtering/ Conditional Filtering

```python
# View all the column-data ("MILES") where miles > 10; returns boolean values; conditional filtering
df['MILES'] > 10
```

```python
# Single Condition
# View all the rows including all columns where the "MILES" column-data > 10
# (Approach-1) df.loc[df['MILES']>10]
df[df['MILES']>10]
```

```python
# Single Condition
# View all the rows of only the "MILES" column where the "MILES" > 10
# (Approach-1) df.loc[df['MILES']>10, ['MILES']]
# (Approach-2) df.MILES[df['MILES']>10]
df["MILES"][df['MILES']>10]
df["MILES"].isin([10]) # require to pass a list, rather than as a int/str
```

```python
# Single Condition
# View all the rows of only the "MILES" & "PURPOSE" columns where the "MILES" > 10
# (Approach-1) df.loc[df["MILES"]>10,["MILES","PURPOSE"]]
df[["MILES","PURPOSE"]][df["MILES"]>10]
```

```python
# Multiple Conditions
# View all the rows of only the "MILES" & "PURPOSE" columns where the "MILES" > 10 & "PURPOSE" == "meeting"
# (Approach-1) df.loc[(df["MILES"]>10) & (df["PURPOSE"]=="Meeting"), ["MILES","PURPOSE"]]
# df[["MILES","PURPOSE"]][(df["MILES"]>10) & (df["PURPOSE"]=="Meeting")]
df[["MILES","PURPOSE"]][(df["MILES"]>10) & (df["PURPOSE"]=="Meeting")]
```

### Dataframe.isin()

> Filter dataframe by multiple values in a column or multiple columns.

```python
# Find all the rides from Cary & Morrisville; in the "START" column
# (Approach-1) df.loc[df["START"].isin(["Cary","Morrisville"])] # Check whether the "START" column contains "Cary" or "Morrisville"
df[df["START"].isin(["Cary","Morrisville"])]
```

```python
# View all the rows of all the columns except "Cary","Morrisville" in the "START" column
df.loc[~df["START"].isin(["Cary","Morrisville"])].head(10)
```

```python
# [ Task-1 ]: Find all the trips with distance > 10 and originating from Cary & Morrisville
# Make a reset index; remove the "index" column lastly
df2 = df[(df["MILES"]>10) & (df["START"].isin(["Cary","Morrisville"]))]
# Required to reset the index for further accessing this filtered dataset using "iloc" or "loc".
df2.reset_index(inplace=True, drop=True) # silently drop the newly created "index" column
df2

# [ NB ]: Not required if the "drop=True" param is included
# df2.drop(columns="index", inplace=True)
# df2
```

```python
# Add a new column with classic codition using "np.where()"
# [ Intuition ] np.where(condition,retrunValue, elseReturnValue)
df["TRIP_CAT"] = np.where(df["MILES"]>10,"LONG TRIP","SHORT TRIP")
```

### [ Task-2 ]: Create a new column with the following conditions:
+ trip >10 --> Long Trip
+ trip (5-10) --> Medium Trip
+ trip <5 --> Short Trip

2 Appraoches
-------------------
+ **df.apply() - apply a lambda func with the python shorthand conditionals**
+ **np.where() - 2 np.where() funcs in a nested manner**

```python
# Drop the "TRIP_CAT" column
# df.drop(columns="TRIP_CAT", inplace=True)
```

```python
# (Appraoch-1) Create a new conditional column using the "df.apply()" func
# Python shorthand conditional statement is used; if...elif...else
df["TRIP_CAT"] = df["MILES"].apply(lambda x: "Long Trip" if x>10 else "Short Trip" if x<5 else "Medium Trip")
```

```python
# Add a new column with classic codition using "np.where()"
# (Appraoch-2) Create a new conditional column using the "np.where()" func
# [ Intuition ] np.where(condition,retrunValue, np.where(condition, returnValue))
df["TRIP_CAT_(np.where)"] = np.where(df["MILES"]>10,"LONG TRIP",
                                    np.where(df["MILES"]<5,"SHORT TRIP","MEDIUM TRIP"))
```

### Grouping

> [ **NB** ]: In statistical term, grouping is known as SAC (Split, apply, combine)

Ref:
-----
+ [**Pandas Groupby Sort within Groups**](https://sparkbyexamples.com/pandas/pandas-groupby-sort-within-groups/)
+ [**pandas.core.groupby.DataFrameGroupBy.agg**](https://pandas.pydata.org/pandas-docs/version/0.22/generated/pandas.core.groupby.DataFrameGroupBy.agg.html)
+ [**pandas.DataFrame.groupby**](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html)

Syntax: 
+ (Method-1: without "agg()"): dataframe.groupby(groupbyColumn)[ApplyFuncColumn-Optional]
+ (Method-2: with "agg()"): dataframe.groupby(groupbyColumn)[ApplyFuncColumn-Optional].agg()
+ (Method-3: with multiple funcs inside "agg()"): dataframe.groupby('colName').agg({'colName1':'func1', 'colName2':'func2', 'colName':['func3','func4']})

#### Groupby on single column

```python
# View the average miles-trip covered from the start-location
df.groupby("START")["MILES"].agg("mean")
```

```python
df.groupby("START").mean() # It calculates the mean value of "MILES" column
```

```python
# View the average miles-trip covered from the start-location, lastly organize in descending order
df.groupby("START")["MILES"].agg("mean").sort_values(ascending=False)
```

```python
# View the minimum & maximum miles-trip completed from the start-location
df.groupby("START")["MILES"].agg(["min","max"])
```

```python
# View the total miles covered based on the start-location
df.groupby("START")["MILES"].agg("sum")
```

```python
# grouped by the "START" column; apply sum() func on each "MILES" column; 
# then assign new int-index using the 'reset_index()' func; 
# lastly sort the dataframe-rows on "MILES" column in descending order to find the max-aggregated miles-value
# Ref: https://stackoverflow.com/a/36316186
df.groupby("START")["MILES"].agg("sum").reset_index().sort_values(by="MILES", ascending=False)
```

```python
df[df["START"] == "Agnew"]
```

#### Groupby on multiple columns

```python
# Groupby on "START" & "STOP" columns;
df.groupby(["START","STOP"]).agg(["mean","sum"]).head(20)
```

### [ Task ]: Find most recent & earliest travelled date & mean distance travelled for each start city

```python
df.head()
```

```python
# groupby on the "START" column; apply the min, max & mean functions accordingly on the "START_DATE","END_DATE","MILES" columns;
# Inside the "agg()" function; use a dictionary where the keys will be the column-names & the values will be the functions respectively.
# Syntax:  dataframe.groupby('colName').agg({'colName1':'func1', 'colName2':'func2', 'colName':['func3','func4']})
df.groupby("START").agg({
    'START_DATE':'min',
    'END_DATE':'max',
    'MILES':'mean'
})
```

```python
# groupby on the "START" column; apply the min, max & mean functions accordingly on the "START_DATE","MILES" columns;
result_groupby = df.groupby("START").agg({
    'START_DATE':['min','max'],
    'MILES':'mean'
})
```

```python
# Create a separate column of integer-index (0....n)
result_groupby.reset_index(inplace=True)
```

```python
result_groupby.columns
```
